```python
import pandas as pd
data = pd.read_csv("D:\B.Tech\CETPA Notes\ML\Cancer-data.csv")
```

```python
data.head()
```

```python
data = data.drop('id',axis =1 )
```

```python
from sklearn.preprocessing import LabelEncoder
k = LabelEncoder()
data['diagnosis'] = k.fit_transform(data['diagnosis'])
```

```python
data.head()
```

```python
data.max()
```

```python
data['area_worst'] = data['area_worst']/4254
data['area_mean'] = data['area_mean']/2501
data['area_se'] = data['area_se']/542.2
```

```python
data.head()
```

```python
data.max()
```

```python
x = data.drop('diagnosis',axis=1)
y=data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
def model(x,y):
    from sklearn.neighbors import KNeighborsClassifier
    knn = KNeighborsClassifier(n_neighbors=3)
    knn.fit(x,y) 
    from sklearn.naive_bayes import GaussianNB
    gnb = GaussianNB()
    gnb.fit(x,y)
    from sklearn.linear_model import LogisticRegression 
    logistic = LogisticRegression()
    logistic.fit(x,y)
    from sklearn.svm import SVC # "Support vector classifier"
    svc = SVC(kernel='linear', C=1E10)
    svc.fit(x, y)
    return knn,gnb,logistic,svc
```

```python
knn,gnb,logistic,svc = model(x_test,y_test)
```

```python
def accuracy(knn,nb,logistic,svc,x,y):
    a = knn.score(x_test,y_test)
    print('KNN Score: ',a)
    b = gnb.score(x_test,y_test)
    print('Naive Bayes Score: ',b)
    c = logistic.score(x_test,y_test)
    print('Logistic Score: ',c)
    d = svc.score(x_test,y_test)
    print('SVM Score: ',d)
    if a>b and a>c and a>d:
        print('KNN is better')
        return knn
    elif b>c and b>d:
        print('Naive Bayes is better')
        return gnb
    elif c>d:
        print('Logistic is better')
        return logistic
    else:
        print('SVM is better')
        return svc
```

```python
model = accuracy(knn,gnb,logistic,svc,x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
